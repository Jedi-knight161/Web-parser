```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
import asyncio
```

```python
def scrapping_pages(page):
    #html=open('/Users/Andrey/Documents/automobiles1.html','r',encoding='utf-8')
    #url = 'https://auto.ru/moskva/cars/all/?page={}&sort=fresh_relevance_1-desc&special=true&output_type=list'.format(page)
    url = 'https://auto.ru/moskva/cars/all/?sort=fresh_relevance_1-desc&page={}'.format(page)
    html = requests.get(url)
    html.encoding='utf-8'
    #r=BeautifulSoup(html,'lxml')
    r=BeautifulSoup(html.text,'lxml')
    ads= r.find_all('div',{'class':'ListingItem-module__description'})
    d=[]
    for el in ads:
        t0=''
        km_age = el.find('div',{'class':'ListingItem-module__kmAge'}).text
        for i in km_age:
            if i.isdigit():
                t0+=i
            else:
                pass
        km_age=t0
        t1=''
        car_price =  el.find('div',{'class':'ListingItemPrice-module__content'}).text
        car_price = car_price.replace(u'\xa0', u'')
        car_price =  car_price.replace('₽','')
        for i in car_price:
            if i.isdigit():
                t1+=i
            else:
                pass
        car_price=int(t1)
        ad_link = el.find('a',{'class':'Link ListingItemTitle-module__link'}).get('href')
        car_name = el.find('a',{'class':'Link ListingItemTitle-module__link'}).get_text()
        descr= el.find('div',{'class':'ListingItemTechSummary-module__column'}).parent('div',{'class':'ListingItemTechSummary-module__cell'})
        year = el.find('div',{'class':'ListingItem-module__year'}).text
        bodywork = descr[2].text
        drive = descr[3].text
        color = descr[4].text
        trans_type =descr[1].text
        engine = descr[0].text
        engine =  engine.replace(u'\u2009','')
        engine =  engine.replace(u'\xa0', '')
        engine_power =  engine.split(u'/')[1]
        engine_volume = engine.split(u'/')[0]
        engine_type =  engine.split(u'/')[2]
        d.append([ad_link,car_name,year,km_age,bodywork,drive,trans_type,engine_power,engine_volume,engine_type,car_price])
    return d
```

```python
scrapping_pages(1)
```

```python
#df = pd.DataFrame(data=[[]],columns=['Ссылка','Модель','Год выпуска','Пробег','Тип кузова','Привод','Трансмиссия','Мощность двиателя','Объём двигателя','Тип двигателя','Цена'])
dat=[]
#i=0
#a=1
for i in range(1000):
    dat+=scrapping_pages(i)
#while a!=[]:
    #a=scrapping_pages(i)
    #dat+=a
    #i+=1 
```
